Implement a function printRightTriangle(  height ) which prints an ascii triangle (using '*'s) of height height.

Theory

Derive a step count function T(h) where h is height. Note that h is the value of the input and not the size of the input, thus we will not do a best, worst, and average case analysis. 
Determine a tight-fit upperbound of T(h) using Big-O notation.

Practical

Using the time module, measure the runtime of printRightTriangle(h) and plot the relationship between the runtime and h: h vs. runtime_in_seconds.
Run multiple simulations at a variety of values of h to confirm your derivation from Theory Section above. 
Check out code example which uses time module to track runtime of algorithms

Present and demo your code and resulting figures to the class as a big group. 

```python
import time
import matplotlib.pyplot as plt
```

```python
def printRightTriangle(height):
    starttime = time.time()
    for i in range(1, height+1):
        for j in range (i):
            print("*", end = "")
        print("\n")
    endtime = time.time()
    runtime = endtime - starttime
    return runtime
```

```python
# for testing
printRightTriangle(3)
```

```python
import numpy as np
heights = np.arange(1,200,5)
timets = [printRightTriangle(height) for height in heights]

plt.plot(heights,timets)
plt.xlabel('Value of height')
plt.ylabel('Running time')
plt.title('Height VS Runtime_in_seconds.')
plt.show()
```

```python
# T(h) is O(h^2)

def T(h):
    return(2*h+pow(2*h,2)+pow(h,2)+h)
```

```python
import numpy as np
hs = np.arange(1,200,5)
times = [T(h) for h in hs]
plt.plot(hs, times)
plt.xlabel('Value of height')
plt.ylabel('Running time')
plt.title('Height VS Runtime_in_seconds.')
plt.show()  #seems quadratic -- O(n^2)
```
